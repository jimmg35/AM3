```python
import numpy as np
import pandas as pd
from os import listdir
import os
```

```python
data = pd.read_csv(r"./education/S1502_Field_of_Bach/ACSST5Y2023.S1502-Data.csv", header=None)
data = data.drop(0, axis=0)
data.columns = data.iloc[0]
data = data.drop(data.index[0])
data = data.reset_index(drop=True)
```

```python
cols = [
    'Geography',
    'Geographic Area Name',
    "Estimate!!Total!!Total population 25 years and over with a Bachelor's degree or higher"
]

# keep only columns that exist (avoids KeyError) and create a copy
missing = [c for c in cols if c not in data.columns]
if missing:
    print("Missing columns (will be skipped):", missing)

new_df = data[[c for c in cols if c in data.columns]].copy()
```

```python

new_df.rename(columns={
    "Estimate!!Total!!Total population 25 years and over with a Bachelor's degree or higher": 'population_25p_bachelors_or_higher'
}, inplace=True)

new_df.head()
output_path = "2023_ACS_5years_education_census_tract_results.csv"
new_df.to_csv(output_path, index=False, encoding="utf-8")
print(f"Saved to {output_path}")
```
