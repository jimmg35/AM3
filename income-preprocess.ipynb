```python
import numpy as np
import pandas as pd
from os import listdir
import os
```

```python
data = pd.read_csv(r"./income/S1901_Income in the past 12 m/ACSST5Y2023.S1903-Data.csv", header=None)
data = data.drop(0, axis=0)
data.columns = data.iloc[0]
data = data.drop(data.index[0])
data = data.reset_index(drop=True)

# 
```

```python
cols = [
    'Geography',
    'Geographic Area Name',
    "Estimate!!Median income (dollars)!!HOUSEHOLD INCOME BY RACE AND HISPANIC OR LATINO ORIGIN OF HOUSEHOLDER!!Households"
]

# keep only columns that exist (avoids KeyError) and create a copy
missing = [c for c in cols if c not in data.columns]
if missing:
    print("Missing columns (will be skipped):", missing)

new_df = data[[c for c in cols if c in data.columns]].copy()
```

```python

new_df.rename(columns={
    "Estimate!!Median income (dollars)!!HOUSEHOLD INCOME BY RACE AND HISPANIC OR LATINO ORIGIN OF HOUSEHOLDER!!Households": 'median_household_income'
}, inplace=True)

new_df.head()
output_path = "2023_ACS_5years_income_census_tract_results.csv"
new_df.to_csv(output_path, index=False, encoding="utf-8")
print(f"Saved to {output_path}")
```
