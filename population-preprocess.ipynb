```python
import numpy as np
import pandas as pd
from os import listdir
import os
```

```python
data = pd.read_csv(r"./population/AgeSex_2024_census_tract/ACSST5Y2023.S0101-Data.csv" , header=None)
data = data.drop(0, axis=0)
data.columns = data.iloc[0]
data = data.drop(data.index[0])
data = data.reset_index(drop=True)
```

```python
cols = [
    'Geography',
    'Geographic Area Name',
    'Estimate!!Total!!Total population',
    'Estimate!!Total!!Total population!!SELECTED AGE CATEGORIES!!18 to 24 years',
    'Estimate!!Total!!Total population!!AGE!!25 to 29 years',
    'Estimate!!Total!!Total population!!SELECTED AGE CATEGORIES!!65 years and over',
    'Estimate!!Male!!Total population'
]

# keep only columns that exist (avoids KeyError) and create a copy
missing = [c for c in cols if c not in data.columns]
if missing:
    print("Missing columns (will be skipped):", missing)

new_df = data[[c for c in cols if c in data.columns]].copy()
```

```python
new_df['population_18_29'] = (
    pd.to_numeric(new_df['Estimate!!Total!!Total population!!SELECTED AGE CATEGORIES!!18 to 24 years'], errors='coerce').fillna(0).astype(int)
    + pd.to_numeric(new_df['Estimate!!Total!!Total population!!AGE!!25 to 29 years'], errors='coerce').fillna(0).astype(int)
)
new_df.drop([
    'Estimate!!Total!!Total population!!SELECTED AGE CATEGORIES!!18 to 24 years',
    'Estimate!!Total!!Total population!!AGE!!25 to 29 years'
], axis=1, inplace=True)
new_df.rename(columns={
    'Estimate!!Total!!Total population': 'total_population',
    'Estimate!!Total!!Total population!!SELECTED AGE CATEGORIES!!65 years and over': 'population_65',
    'Estimate!!Male!!Total population': 'population_male'
}, inplace=True)

new_df.head()
output_path = "2023_ACS_5years_population_census_tract_results.csv"
new_df.to_csv(output_path, index=False, encoding="utf-8")
print(f"Saved to {output_path}")


```
