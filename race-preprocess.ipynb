```python
import numpy as np
import pandas as pd
from os import listdir
import os
```

```python
data = pd.read_csv(r"./race/B03002_Hispanic_or_Latino/ACSDT5Y2023.B03002-Data.csv", header=None)
data = data.drop(0, axis=0)
data.columns = data.iloc[0]
data = data.drop(data.index[0])
data = data.reset_index(drop=True)
```

```python
cols = [
    'Geography',
    'Geographic Area Name',
    'Estimate!!Total:!!Not Hispanic or Latino:!!Black or African American alone',
    'Estimate!!Total:!!Not Hispanic or Latino:!!American Indian and Alaska Native alone',
    'Estimate!!Total:!!Not Hispanic or Latino:!!Asian alone',
    'Estimate!!Total:!!Hispanic or Latino:'
]

# keep only columns that exist (avoids KeyError) and create a copy
missing = [c for c in cols if c not in data.columns]
if missing:
    print("Missing columns (will be skipped):", missing)

new_df = data[[c for c in cols if c in data.columns]].copy()
```

```python

new_df.rename(columns={
    'Estimate!!Total:!!Not Hispanic or Latino:!!Black or African American alone': 'population_black',
    'Estimate!!Total:!!Not Hispanic or Latino:!!American Indian and Alaska Native alone': 'population_american_indian',
    'Estimate!!Total:!!Not Hispanic or Latino:!!Asian alone': 'population_asian',
    'Estimate!!Total:!!Hispanic or Latino:': 'population_hispanic'
}, inplace=True)

new_df.head()
output_path = "2023_ACS_5years_race_census_tract_results.csv"
new_df.to_csv(output_path, index=False, encoding="utf-8")
print(f"Saved to {output_path}")
```
