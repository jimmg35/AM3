```python
import pandas as pd
```

```python
df = pd.read_csv(r"./smoking/PLACES__Census_Tract_Data_(GIS_Friendly_Format),_2024_release_20251114.csv")
df = df[df['StateAbbr'] == 'FL'].reset_index(drop=True)
df.head()
```

```python
cols = [
    'CountyName',
    'CountyFIPS',
    'TractFIPS',
    'CSMOKING_CrudePrev'
]

# keep only columns that exist (avoids KeyError) and create a copy
missing = [c for c in cols if c not in df.columns]
if missing:
    print("Missing columns (will be skipped):", missing)

new_df = df[[c for c in cols if c in df.columns]].copy()
```

```python

new_df.rename(columns={
    "CSMOKING_CrudePrev": 'smoking_crude_prevalence'
}, inplace=True)

new_df.head()
output_path = "2024_CDC_PLACES_smoking_census_tract_results.csv"
new_df.to_csv(output_path, index=False, encoding="utf-8")
print(f"Saved to {output_path}")
```
